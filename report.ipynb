 <p style="font-family:Georgia; font-size:18px"><b>1.</b></p> 

\begin{equation}
\begin{aligned}
L\left(\mu \mid x_{1}, x_{2}, \ldots, x_{N}\right) &=\prod_{i=1}^{N} L\left(\mu \mid x_{i}\right) \\
&=\prod_{i=1}^{N} \frac{1}{\sqrt{2 \pi \sigma^{2}}} \cdot e^{-\frac{\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}}
\end{aligned}
\end{equation}

<p style="font-family:Times New Roman; font-size:17px">We take the log of  the likelihood function to make the derivation easy (since log is a monotonically increasing function, we can make use of it):</p>

\begin{aligned}
\ln \left(L\left(\mu \mid x_{1}, x_{2}, \ldots, x_{n}\right)\right) &=\sum_{i=1}^{N}\left[\ln \left(\frac{1}{\sqrt{2 \pi \sigma^{2}}}\right)+\ln \left(e^{-\left(x_{i}-\mu\right)^{2} / 2 \sigma^{2}}\right)\right] \\
&=\sum_{i=1}^{N}\left[\ln \left(2 \pi \sigma^{2}\right)^{-1 / 2}-\frac{\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}  \ln (e)\right] \\
&=\sum_{r=1}^{N}\left[-\frac{1}{2} \cdot \ln (2 \pi)-\frac{1}{2} \ln \left(\sigma^{2}\right)-\frac{\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}\right] \\
&=\sum_{i=1}^{N}\left[-\frac{1}{2} \ln (2 \pi)-\ln (\sigma)-\frac{\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}\right] \\
&=\left[-\frac{N}{2} \cdot \ln (2 \pi)-N  \ln (\sigma)+\sum_{i=1}^{N} \frac{-\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}\right]
\end{aligned}

<p style="font-family:Times New Roman; font-size:17px">We take the partial derivative of the equation above w.r.t $\mu$ and equating to zero to estimate the mean value: </p>

\begin{equation}
\begin{aligned}
\frac{\partial}{\partial \mu} \ln \left(L\left(\mu \mid x_{1}, x_{2}, \ldots, x_{N}\right)\right)=& 0-0+\sum_{i=1}^{N} \frac{2\left(x_{i}-\mu\right)}{2 \sigma^{2}} \\
=\frac{1}{\sigma^{2}}\left[\sum_{i=1}^{N} x_{i}-N \cdot \mu\right]=0 \\
N \mu=\sum_{i=1}^{N} x_{i} \\
\mu=\frac{\sum_{i=1}^{N} x_{i}}{N}
\end{aligned}
\end{equation}

\begin{equation}
\begin{aligned}
L &=\prod_{n=1}^{N}\left[P\left(c^{n}\right) \cdot \prod_{i=1}^{D} P\left(x_{i}^{n} \mid c^{n}\right)\right] \\
\ln (L) &=\sum_{n=1}^{N} \ln \left(P\left(c^{n}\right) \cdot \prod_{i=1}^{D} P\left(x_{i}^{n} \mid c^{n}\right)\right) \\
&=\sum_{n=1}^{N}\left[\ln \left(P\left(c^{n}\right)\right)+\ln \left(\prod_{i=1}^{D} P\left(x_{i}^{n} \mid c^{n}\right)\right)\right] \\
&=\sum_{n=1}^{N}\left[\ln \left(P\left(c^{n}\right)\right)+\sum_{i=1}^{D} \ln \left(P\left(x_{i}^{n} \mid c^{n}\right)\right)\right]
\end{aligned}
\end{equation}

<p style="font-family:Times New Roman; font-size:17px">Since we have binary attributes, we have bernoulli distribution. So we replace $P(x_{i}^{n} \mid c^{n})$ with equation below:</p>

\begin{equation}
\begin{aligned}
&=\sum_{n=1}^{N}\left[\ln (P(c=0))+\ln (P(c=1))+\sum_{i=1}^{D} \ln \left(\left(\theta_{i}^{c^{n}}\right)^{x_{i}^{n}}\left(1-\theta_{i}^{c^{n}}\right)^{\left(1-x_{i}^{n}\right)}\right)\right] \\
=& \sum_{n=1}^{N}\left[\ln (P(c=0))+\ln (P(c=1))+\sum_{i=1}^{D}\left[x_{i}^{n} \ln \left(\theta_{i}^{c^{n}}\right)+\left(1-x_{i}^{n}\right) \cdot \ln \left(1-\theta_{i}^{c^{n}}\right)\right]\right.\\
=& n_{0}  \ln (P(c=0))+n_{1}  \ln (P(c=1))+\sum_{n=1}^{N} \sum_{i=1}^{D}\left[x_{i}^{n} \ln \left(\theta_{i}^{c^{n}}\right)+\left(1-x_{i}^{n}\right) \ln \left(1-\theta_{i}^{c^{n}}\right)\right]
\end{aligned}
\end{equation}

<p style="font-family:Times New Roman; font-size:17px">Then we use indicator functions of the event $x_{i}^{n} = {0,1}$:</p>

\begin{equation}
\begin{aligned}
=\sum_{i, n}\left[\mathbb{I}\left[x_{i}^{n}=1, c^{n}=c\right] \ln \left(\theta_{i}^{c^{n}}\right)\right.&\left.+ \mathbb{I}\left[x_{i}^{n}=0, c^{n}=c\right] \cdot \ln \left(1-\theta_{i}^{c^{n}}\right)\right] +n_{0}  \ln (P(c=0))+n_{1} \ln (P(c=1))
\end{aligned}
\end{equation}

<p style="font-family:Times New Roman; font-size:17px">After that we take the partial derivative of the above equation w.r.t $\theta_{i}^{c^{n}}$ :</p>


\begin{equation}
\frac{\partial}{\partial\left(\theta_{i}^{c}\right)} \ln (L)=\sum_{i, n}\left[\frac{\mathbb{I}\left[x_{i}^{n}=1, c^{n}=c\right)}{\theta_{i}^{c^{n}}}-\frac{\mathbb{I}\left[x_{i}^{n}=0, c^{n}=c\right]}{\left(1-\theta_{i}^{c^{n}}\right)}\right]+0+0=0
\end{equation}

\begin{equation}
\begin{array}{l}
\sum_{i, n} \frac{\left(1-\theta_{i}^{c^{n}}\right) \mathbb{I}\left[x_{i}^{n}=1, c^{n}=c\right]-\theta_{i}^{c^{n}} \mathbb{I}\left[x_{i}^{n}=0, c^{n}=c\right]}{\theta_{i}^{c^{n}}\left(1-\theta_{i}^{c^{n}}\right)}=0 \\
\sum_{i, n} \frac{\mathbb{I}\left[x_{i}^{n}=1, c^{n}=c\right]-\theta_{i}^{c^{n}} \left(\mathbb{I}\left[x_{i}^{n}=1, c^{n}=c\right]+\mathbb{I}\left[x_{i}^{n}=0, c^{n}=c\right]\right)}{\theta_{i}^{c^{n}}\left(1-\theta_{i}^{c^{n}}\right)}=0 \\
\theta_{i}^{c}=\frac{\sum_{i, n} \mathbb{I}\left[x_{i}^{n}=1, c^{n}=c\right]}{\sum_{i, n}\left[\mathbb{I}\left[x_{i}^{n}=1, c^{n}=c\right]+I\left[x_{i}^{n}=0, c^{n}=c\right]\right]}
\end{array}
\end{equation}

<p style="font-family:Times New Roman; font-size:17px">In equation above, numerator part indicates the number of times $x_{i} = 1$ for class $c$ and denominator part indicates number of datapoints in class $c$</p>

<p style="font-family:Georgia; font-size:23px"><b>Part II: Sentiment Analysis with Naive Bayes</b></p> 

<p style="font-family:Times New Roman; font-size:18px">In this assignment, we implement a naive bayes classifier using the bag of words(BoW) representation with unigram and bigram options and use it to try determining whether a customer review is positive or negative. We also extend our classification task to predicting cateagories(books, camera, dvd, health, music, or software) of the given review. We will also use different techniques to improve our model performance such that removing stop words and adapting tf-idf weights into naive bayes classification problem.</p> 

<p style="font-family:Times New Roman; font-size:17px; text-align:center"><b>Data Analysis</b></p> 

<p style="font-family:Times New Roman; font-size:18px">We see that we have balanced traning data, since the training data samples are distributed equally in terms of the number of review for each class of sentiment feature and for each class of category feature. </p> 
    
<p style="font-family:Times New Roman; font-size:18px">Since we use BoW model, we want that each sentiment should be represented by dictionaries whose sum of word frequencies are equal and they should include variety of unique words that reflects characteristics of its sentiment. So, the number of words used in reviews are basically more important than the number of reviews when we decide if we use a balanced dataset, since a review can be just a few words or just a few paragraphs long. Accordingly, we observe that the distribution of number of words used in a review are balanced based on their sentiments. In addition, each category (music, books, dvd, camera, health, software) has equally distributed reviews for each sentiment feature.</p>         

<p style="font-family:Times New Roman; font-size:17px; text-align:center"><b>Preview on Sentiments: The distribution of review sentiment polarity score</b></p> 

<p style="font-family:Times New Roman; font-size:17px; text-align:center"><b>Preview on Sentiments: Example sentences according their polarity scores</b></p> 

<p style="font-family:Times New Roman; font-size:18px">We also wanted to use polarity scores to get an intuition and see examples of highly positive reviews which have high polarity score, highly negative reviews which have low polarity score and neutral reviews which have 0 polarity score. Polarity scores is in range [-1,1]. -1 means highest negative sentiment score and 1 means highest positive sentiment score. You can see the review examples above based on their score. We can clearly see that some keywords dominate the review's polarity score and their sentiment class accordingly. "best, happy, excellent, impressed, perfectly ..." are the words used frequently in highly positive sentiments, while "boring, worst, talentless, terrible, horrible, hated ..." are the words used frequently in highly negative sentiments. When it comes to neutral reviews, they generally includes both words used in negative sentiments and positive sentiments. And they are longer reviews comparison with negative and positive reviews. Since they are longer, they include more words. The more words the more emotions, the more expressions. Thus, they can include characteristics words in both sentimen classes, and so that makes them neutral.</p>

<p style="font-family:Times New Roman; font-size:18px">We use different settings to get better accuracy on our model. We use BoW with three options: unigram, bigram, trigram. We have two option for review which are given review and cleared review. Cleared review means that we remove punctuations and numbers from each review in dataset. We also try our model by removing stopwords from reviews. In addtion, we try to observe how the accuracy will change, if we remove stopword from the model but not from the test set and vice versa. In the end, we get our best model with <b>86.32%</b> accuracy via bigram BoW model, non-cleared reviews and stopwords are included. </p> 

<p style="font-family:Times New Roman; font-size:18px">To implement TF-IDF in terms of the the conditional probabilities used in the Naive Bayes algorithm, we use tf-idf weight of the word, which is obtained by multiplying frequency of word with idf value of the word, instead of frequency of the word in the class and use sum of tf-idf weights of the class instead of total word frequency of the class. IDF values are obtained by getting the log of the total number of documents belonging the class over how many documents include that word. However, we get lower accuracy score (<b>73.3%</b>) than the score we get from the model we use only frequencies.</p>

<p style="font-family:Times New Roman; font-size:18px"> There can be several reasons why the model misclassifies the reviews. We list a following possible reasons and give corresponding examples above:</p> 

<ul style="font-family:Times New Roman; font-size:18px">
  <li> (1) Since we use bigram option for the BoW model, we take successive words when creating the model. There are many verbs with negations (e.g. n't). Although the overall sentiment is positive, those negations may make the model predict the sentiment as negative. </li>
  <li> (2) There are many noun and adjectives,which have characteristics of negative sentiment, in the review; although it's a positive review.</li>
  <li> (3) There is sarcasm in the review and our simple model may not be able to understand this. </li>
  <li> (4) In the review, the user quotes negative comments made by someone else and use negative expressions when criticizing other comments. That may makes the model predict the sentiment as negative, because the model thinks using expressions and words in review belong to user's own opinion.</li>
  <li> (5) While it's a negative comment, there are positive words and expressions in it. In addtion, user actually gives a positive comment for other products in her/his review. </li>
</ul>
